# Multiple Linear Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
#  importing the data set
dataset = pd.read_csv("50_Startups.csv")
# seperating the dependent and independent variables
x = dataset.iloc[ : , :-1].values
y = dataset.iloc[: , -1].values
# print(x)
```

```python
dataset.head()
```

## Encoding categorical data

```python
# Encoding categorical data
# Encoding the Independent Variable
from sklearn.compose import ColumnTransformer # importing the ColumnTransformer class
from sklearn.preprocessing import OneHotEncoder # importing the OneHotEncoder class
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough') # creating an instance of the ColumnTransformer class
x = np.array(ct.fit_transform(x)) # applyin
```

```python
print(x)
```

## Splitting the dataset into the Training set and Test set

```python
# tesplit function for splitinf the data into test and training sets
from sklearn.model_selection import train_test_split
# split the data sets into training and testing sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

## Training the Multiple Linear Regression model on the Training set

```python
# import the Linear Regression class
from sklearn.linear_model import LinearRegression
# create the linear regression object
regressor = LinearRegression()
# train our object with the dataset 
regressor.fit(x_train, y_train) 
```

## Predicting the Test set results

```python
#  predict the response of the test set based on our trained moddel
y_predict = regressor.predict(x_test)
# convert all numerical values in 2 decimal places with numpy
np.set_printoptions(precision=2)
# creat a numpy concatenation of the two arrays
print(np.concatenate((y_test.reshape(len(y_test),1), y_predict.reshape(len(y_predict),1)),1))
```

```python
# create a dataframe to view our prediction and the actual array in the test set
prediction_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_predict})
prediction_df
```
